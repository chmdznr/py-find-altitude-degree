```python
import matplotlib.pyplot as plt
import pysolar.solar
from datetime import datetime, timedelta
import pytz
```

```python
latitude = -6.9667  # Semarang
longitude = 110.4167
tz = 'Asia/Jakarta'

start_time = datetime(2023, 12, 25, 6, 0, 0, tzinfo=pytz.timezone(tz))
end_time = datetime(2023, 12, 25, 18, 0, 0, tzinfo=pytz.timezone(tz))
```

```python
times = [start_time + timedelta(minutes=30*i) for i in range(int((end_time - start_time).total_seconds() / 1800))]
altitudes = [pysolar.solar.get_altitude(latitude, longitude, time) for time in times]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(times, altitudes)
plt.xlabel('Time of Day')
plt.ylabel('Solar Altitude (degrees)')
plt.title('Solar Altitude throughout the Day in Semarang at ' + str(start_time.date()) + ' (UTC+7)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
